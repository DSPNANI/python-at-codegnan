### Input Formating

```python
# reading integer from user
num = int(input("Enter a number:"))
print(type(num))
print(num)
```

```python
# reading float vlaue from user
val = float(input())
print(type(val))
print(val)
```

```python
# complex reading from user
val = complex(int(input()), int(input()))
print(type(val))
print(val)
```

```python
# boolean reading
val = bool(input())
print(type(val))
print(val)
```

```python
# boolean reading as integer
val = bool (int(input()))
print(type(val))
print(val)
```

```python
# boolean reading as integer
val = bool(int(input()))
print(type(val))
print(val)
```

```python
# reading string value from user
val = input()
print(type(val))
print(val)
```

```python
# reading list of integers from user
lst = list(map(int,input().split()))
print(lst)
```

```python
# reading list of string from user
lst = list(map(str,input().split()))
print(lst)
```

```python
# approach 2
lst = input().split()
print(lst)
```

```python
# reading Tuple of integers from user
lst = tuple(map(int, input().split()))
print(lst)
```

```python
# eval(): this function evalutes the string type extraction
val = eval(input())
print(type(val))
print(val)
```

```python
print("3+5")
print(eval("3+5"))
```

```python
name = input()
age = input()

print(f"My name is {name} and I am {age} years old")
```

```python
name = " Varaprasad"
age = "20"
print("My name is " , name,"and i am " , age,"years old")
```

```python
# Access Specifiers
#string - %s
# integer - %d
# float - %f
name = "dsp"
age="21"
#My name is dsp and i am 21 years old
print("My name is %s and i am %d years old" %(age,name))
```

```python
# 3 dot format
name = "dsp"
age = 21
# My name is dsp and i am 21 years old
print("My name is {} and i am {} years old".format(name,age))
print("My name is {} and i am {} years old".format(age,name))
```

```python
# 3 f-format
name = "dsp"
age = 21
# My name is dsp and i am 21 years old
print(f"My name is {name} and I am {age} years old")
print(f"My name is {age} and I am {name} years old")
```
