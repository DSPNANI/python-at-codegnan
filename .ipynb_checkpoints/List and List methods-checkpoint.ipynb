### List Characterstics

```python
# Define list
#List is odered Collection
lst = [1,2,3,4.5,"Hi","Hello", 1,3,4]
print(lst)
```

```python
lst = [1,3,4.5,"Hi","Hello", 1,3,4]
print(lst[5])
print(lst[1::2])
print(lst[4:7:1])
print(lst[-3:-1])
```

```python
#empty list creation
l1 = []
l2 = list()
print(type(11), type(12))
```

```python
# List operations
# 1. List Concatination (using + operator)
# 2. List Repetition (using * operator)
l1 = [1,2,3]
l2 = [3,4,5,6]
print(l1+l2)
print(l1+l1)
print(l1*3)
print([0] * 10)
```

### List Methods
- Item adding methods
- append(item) : It adds item at end of list
- extend(iterator) : It adds the all iterator element  at end of list
- insert(pos , item) : It adds item at specified position

```python
l = [1]
l.extend((1,2))
l
```

```python
lst = [10,20,30,40]
lst.append(100)
print(lst)
lst.extend([1,2,3])
print(lst)
lst.append([100,200])
print(lst)
lst.insert(0,1000)
print(lst)
```

### Item removing methods
- remove(item): It removes the first occurence items from the list if it is exists in list otherwise 'ValueError'
- pop(pos) : It remove and return the specified position item if position is passed, otherwise it remove and return the last item form  list.
- clear() : It removes all elements from list

```python
lst = [10,20,30,20,40]
(lst.remove(20))
print(lst)
# list.remove
```

```python
lst = [10,20,30,20,40,50,20]
val = lst.pop()
print(val)
print(lst)
val2 = lst.pop(0) #it removes the 0th position
print(val2)
print(lst)
lst.clear()
print(lst)
```

```python
# idex and count()
# index(iem): If item is present in list it returns the First occurence Index otherwise ValueError
# count(item): It return the count of total occurance of item
lst = [10,20,30,20,40,50]
ind = lst.index(20)
print(ind)
print(lst.count(20))
```

```python
# Sorting items
# sort(reverse = False) : By default it sorts the origin list in into Ascending order.if reverse is true it sorts in Decending order
lst = [1,4,2,5,-5,3,-45,-4,30]
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
lst = ['hi' , 'hello','banti']
lst.sort()
print(lst)
lst.sort(reverse = True)
print(lst)
```

```python
#Buiting functions- len()  , min() , max() , sum()
lst  = [1,2,3,4,5]
print(len(lst),min(lst), max(lst), sum(lst))
print(sum(lst) / len(lst))
```
