### sets : sets is an unodered collection of different datatype ,  unique elements sets are mutable and defines with '{ }'

## Characteristics:
- set is unordered collection
- set elements are unique.
- set doesnot supports indexing.
- set are mutable.
- set allows hashable datatytpe as elements.
- set elements can be used as dicitionary keys.

```python
# set is unordered collection
# set elements return  are unique
s = {1,2,3,2,3,'hi','a',2.02, (10,20,30)}
print(s)
# print(s[0]) # set doesnot support indexing
```

```python
# list can't be used as set element 
s= {1,2,[3,2,3],'hi','a',2.02,(10,20,30)}
```
