# Dictinatary : Dicitionary is orderred collection Key - value(i.e items) pairs. It is mutable and defines with "{}"
# dicitionary format: { Key : Value}
# Characteristics :
- Dicitionary is ordered collection
- Dicitionary  is mutable
- Dicitionary doen0t  support index
- Dicitionary keys are unique.
- Dicitionary values allows duplicates
- List canot be used as dicitionary key, because unhashable.
- dicitionary values can access by using keys

```python
details = {'Name' : 'dsp' , 'course' :'python' , 'batch' : 57}
details [0] # code will be error
```

```python
details = {'Name' : 'dsp' , 'course' :'python' , 'batch' : 57}
print(details['Name'])
```

```python
# Dicitionary is ordered collection
details = {'Name' : 'dsp' , 'course' :'python' , 'batch' : 57, 'age' : 21}
print(details)
```

```python
# Dictionary values access bu using keys
# updating dictionary values
# addding new items to dictionary
details = {'name' : 'dsp' , 'course' :'python' , 'batch' : 57, 'age' : 21}
print(details['course'])
# print(details[location]) # because locate is not there in details (KeyError : 'Location)
#updating dictionary values
details['age'] = details ['age'] + 1
print(details['age'])
#adding new items
details['location'] = "hyderabad"
print(details)
```

```python
# dictionary keys are unique
details = {'age' : 20 , 'name' : 'dsp' , 'course' :'python' , 'batch' : 57, 'age' : 21}
print(details)
```

```python
# List can't be used as dictionary key
details = {'age' : 20 , 'name' : 'dsp' , 'course' :'python' , [1,2,3] : 57, 'age' : 21}
print(details)
```

### Dictionary Methods


- get(key, default_value : Optional) : It returns the value if key is exits in dictionary, otherwise it return default value.
  

```python
details = {'name': 'dsp', 'course': 'python', 'batch': 57, 'age': 22, 'location': 'hyderabad'}
print(details['name'])
# print(details['marks')
print(details.get('name'))
print(details.get('marks')) # get method by default it return none  if key is not exist
print(details.get('marks' ,"marks not availble")) # it returns default value
```

- update(new_dictonary) : This method updates the values if new_dictionary keys already exists in Dictionary, otherwise it adds new items to dictionary.

```python
details = {'name': 'dsp', 'course': 'python', 'batch': 57, 'age': 22}
details.update({'batch' : 58 , 'course' : 'java' , 'location' : 'hyd'})
print(details)
```

# Keys() : The method returns the list dictionary keys
# Values() : This method returns the list dictionary values
# items(): This method returns the List of tuple of Key-value pairs

```python
print(details.keys())
print(details.values())
print(details.items())
```

### Dictionary remove methods
