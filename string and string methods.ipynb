## String Methods (13-06-2026)
- Case Conversion Methods
    - upper(): It converts entire string into upper case
    - lower(): It converts entire string into Lower case
    - title(): It converts the first letter of every word into upper and remain lower
    - swapcase(): It converts upper case letter into lower case and lower to upper
    - capitalize(): It converts First character into upper case and remaining into lower case

```python
string = "The Sky is blue and Ocean is Blue"
print(string.upper())
print(string.lower())
print(string.title())
print(string.swapcase())
print(string.capitalize())
```

- Remove white space or pattern at ends of string 
    - strip(): By default it removes the white space at both ends of string
    - lstrip(): It removes the white space at left side of string
    - -rstrip(): It removes the white space at right side of string

```python
s = "         Codegnan        "
print(s)
print(s.strip())
print(s.lstrip())
print(s.rstrip())
```

```python
s = "@@@@@@@Code@@@gnan@@@@@@@@"
print(s)
print(s.strip("@"))
print(s.lstrip("@"))
print(s.rstrip("@"))
```

```python
# split(): It splits the  entrire string into list of substring based on speccified seperator
# by default it splits at splits at space
string = " The Sky is blue and Ocean is Blue"
print(string.split())
print(string.split("is")) # is will be remove from sentence
print(string.split('e'))
```

```python
# 'sep'.join(iterable): It joins the all iterable elements with seperator
words =  ['The', 'Sky', 'is', 'blue', 'and', 'Ocean', 'is', 'Blue']
print("".join(words))
print(" ".join(words))
print("-".join(words))
```

### Substring Index finding
-index(substring): If substrings is exits in string , it returns the first occurencce of substring position otherwise 'ValueError'
    -find(substring): If substrings is exits in string , it returns the first occurencce of substring position otherwise '-1'

```python
string = "The Sky is blue and Ocean is Blue"
print(string.index('is'))
print(string.index('z'))
```

```python
string = "The Sky is blue and Ocean is Blue"
print(string.find('is'))
print(string.find('z'))
print(string.find('is',10)) # it starts checking form 10th position
print(string.find('e',10))
print(string.find('e', -5))
print(string.find('z',1))
```

### String checking methods
- String checking methods returns the boolean value as result
- isupper(): It return True, If entire string is in upper case
- islower(): It return True, If entire string is in lower case
- isalpha(): It return True, If entire string is in contains alphabets
- isdigit(): It return True, If entire string contains digits
- isalnum(): It return True, If entire string contains either alphabets or digits
- startswith(substring): It return True, If  string starts with substring
- endswith(substring): It return True, If  string ends with substring 

```python
print("CODE".isupper())
print("CODE".islower())
print("code".isupper())
print("1234".isdigit())
print('123abc'.isdigit())
print("Code".isalpha()) # alpha means upper case and lower case
print("Code123".isalnum()) # it checks wheather alphabets or numbers
print('1234'.isalnum())
print("code".startswith('c'))
```
