```python
#slicing
s="codegnan"
print(s[2:5])
print(s[2:5:2])
print(s[-4:])
print(s[5:2])
print(s[3:100])
print(s[7:2])
print(s[-1])
print(s[-1:])
print(s[-5:])
print(s[::-1])
print(s[::-2])
print(s[-3::-1])
print(s[-3::1])
```
