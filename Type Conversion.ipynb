```python
# integer to float,boolean,string
num = 10
print(float(num))
print(bool(num))
print(str(num))
print(complex(num))
```

```python
# if val is float
s = "1234.212"
#print(int(s)) # error literal not int
float_val = float(s)
print(float(s))
print(int(float_val))
print(bool(s))
print(list(s))
print(tuple(s))
print(set(s))
```

```python
# List to str, tuple,set
ls = ['a','b','c','d']
print("".join(ls))
print(tuple(ls))
print(set(ls))
```

```python
# tuple to str,set,list
t = ('a','b','c','d')
print("".join(t))
print(list(t))
print(set(t))
```

```python
# dict to str, list,set,tuple
d = {'a':1,'b':2,'c':3}

print(str(d))
print(list(d))          # keys only
print(tuple(d))         # keys only
print(set(d))           # keys only
print(bool(d))

# values conversion
print(list(d.values()))
print(tuple(d.values()))
print(set(d.values()))
```
